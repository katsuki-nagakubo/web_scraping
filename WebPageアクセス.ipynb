```python
#ここでデータの形を確認
#https://qiita.com/n_oshiumi/items/5adf4dcd2219e86d400b

import requests
from bs4 import BeautifulSoup

r = requests.get('https://www.ted.com/talks?language=ja') 
data = BeautifulSoup(r.text, 'html.parser') # 書き込みたいデータ
with open('data.html', mode='w', encoding = 'utf-8') as fw:
   fw.write(data.prettify()) # data.htmlに書き込みモードで「data」を整形して上書きします！
data.html
```

```python
#まとめ①
#TEDの作者の名前をFileに書き出す

import requests
from bs4 import BeautifulSoup
with open('sample.csv','w') as f:
    
    r = requests.get('https://www.ted.com/talks?language=ja') 
    data = BeautifulSoup(r.text, 'html.parser')
    elems = data.find_all("h4", class_="h12 talk-link__speaker")
    for e in elems:
        print(e.getText())
        f.write('{0}\n'.format(e.getText()))
```
